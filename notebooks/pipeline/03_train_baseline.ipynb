# 03 - Baseline Training (LR + XGBoost)
Notebook wrapper for src/train_baseline.py.

Runs training and inspects latest MLflow metrics for the notebook experiment.

```python
from pathlib import Path
import os
import sys

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
os.chdir(PROJECT_ROOT)
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

print(f'Using project root: {PROJECT_ROOT}')
```

```python
from falabella_risk.models.train_baseline import run_training

run_training(
    feature_path=Path('data/processed/features.parquet'),
    model_out=Path('models/baseline_xgb.pkl'),
    random_state=42,
    mlflow_uri=None,
    experiment_name='falabella_baseline_notebook',
)
```

```python
import mlflow

exp = mlflow.get_experiment_by_name('falabella_baseline_notebook')
if exp is None:
    print('Experiment not found yet.')
else:
    runs = mlflow.search_runs([exp.experiment_id], order_by=['start_time DESC'])
    display_cols = [
        'run_id',
        'metrics.lr_auc', 'metrics.lr_pr_auc', 'metrics.lr_f1', 'metrics.lr_brier', 'metrics.lr_latency_ms_per_row',
        'metrics.xgb_auc', 'metrics.xgb_pr_auc', 'metrics.xgb_f1', 'metrics.xgb_brier', 'metrics.xgb_latency_ms_per_row',
    ]
    existing = [c for c in display_cols if c in runs.columns]
    runs[existing].head(1)
```

```python
model_path = Path('models/baseline_xgb.pkl')
print('Baseline model exists:', model_path.exists())
if model_path.exists():
    print('Model size MB:', round(model_path.stat().st_size / (1024 * 1024), 2))
```
